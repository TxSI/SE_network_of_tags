<h1>Betweenness and Weighted Degree vs Popularity-Change</h1>


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('NeTags_popularity_change/full-weighted-indegree-graph-stats.csv')

df.head()
```

```python
class_colors = plt.cm.rainbow(np.linspace(0,1,len(df['0'].unique())))
color_dict = {group:color for group, color in zip(df['0'].unique(), class_colors)}

colors = [color_dict[group] for group in df['0'].values]

x = np.log10(df['1'].values)
y = np.log10(df['betweenesscentrality'].values+1)
# Plot
plt.figure(figsize=(15,10))

plt.scatter(x, y, c=colors, alpha=1)
plt.xlabel('Change(log)')
plt.ylabel('Betweeness Centrality(log)')
plt.show()
```

```python
class_colors = plt.cm.rainbow(np.linspace(0,1,len(df['0'].unique())))
color_dict = {group:color for group, color in zip(df['0'].unique(), class_colors)}

colors = [color_dict[group] for group in df['0'].values]

x = np.log10(df['1'].values-100)
y = np.log10(df['betweenesscentrality'].values+1)
# Plot
plt.figure(figsize=(15,10))

plt.scatter(x, y, c=colors, alpha=1)
plt.xlabel('Change(log)')
plt.ylabel('Betweeness Centrality(log)')
plt.show()
```
